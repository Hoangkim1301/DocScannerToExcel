```python
import cv2
import numpy as np
from PIL import Image

def detect_column_boundaries(image_path, num_columns):
    # Load the image
    image = Image.open(image_path)

    # Convert the image to grayscale
    gray_image = image.convert("L")

    # Convert to numpy array for OpenCV processing
    np_image = np.array(gray_image)

    # Apply Canny edge detection
    edges = cv2.Canny(np_image, 50, 150)

    # Find contours (potential column boundaries)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Sort contours by X-coordinate (left to right)
    sorted_contours = sorted(contours, key=lambda c: cv2.boundingRect(c)[0])

    # Calculate column boundaries
    column_boundaries = [cv2.boundingRect(c)[0] for c in sorted_contours]
    print("Detected column boundaries:", column_boundaries)

    # Crop each column
    for i in range(num_columns):
        x_start = column_boundaries[i]
        x_end = column_boundaries[i + 1] #if i < num_columns - 1 else image.width
        column_image = image.crop((x_start, 0, x_end, image.height))
        column_image.save(f"column{i + 1}.jpg")
        print(f"Saved column{i + 1}.jpg")

# Example usage: Detect column boundaries and crop the image into 5 columns
input_image_path = r"..\documents\Cut.jpg"
num_columns = 5
detect_column_boundaries(input_image_path, num_columns)

```

```python
import cv2
from PIL import Image

# Load image, convert to grayscale, Otsu's threshold
image = cv2.imread(r"..\documents\Cut.jpg")
#print('Image width is: ',Image.open(r"..\documents\ScanTest.jpg").width)    
result = image.copy()
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
#show the image
cv2.imwrite("binary_image.jpg", thresh)

# Detect horizontal lines
horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (800,1))
detect_horizontal = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)
hor_cnts = cv2.findContours(detect_horizontal, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
hor_cnts = hor_cnts[0] if len(hor_cnts) == 2 else hor_cnts[1]
for c in hor_cnts:
    cv2.drawContours(result, [c], -1, (36,255,12), 2)
    cv2.imwrite("result.jpg", result)
    
# Detect vertical lines
vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1,25))
detect_vertical = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, vertical_kernel, iterations=2)
ver_cnts = cv2.findContours(detect_vertical, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
ver_cnts = ver_cnts[0] if len(ver_cnts) == 2 else ver_cnts[1]
for c in ver_cnts:
    cv2.drawContours(result, [c], -1, (36,255,12), 2)
    print('c:',[c])
    cv2.imwrite("result.jpg", result)
    
intersection_points = []
for h_countour in hor_cnts:
    for v_countour in ver_cnts:
        h_x,h_y,w,h = cv2.boundingRect(h_countour)
        v_x,v_y,w,h = cv2.boundingRect(v_countour)
        intersection_points.append((h_x,v_y))    

# Find the bounding box for the intersection points
x_min = min(point[0] for point in intersection_points)
y_min = min(point[1] for point in intersection_points)
x_max = max(point[0] for point in intersection_points)
y_max = max(point[1] for point in intersection_points)

# Crop the image to the bounding box
trimmed_image = image[y_min:y_max, x_min:x_max]

# Save the trimmed image
#cv2.imwrite("trimmed_image.jpg", trimmed_image)

# Print a success message
#print("ver_cnts :", ver_cnts)
print("Trimmed image saved as trimmed_image.jpg")

```
